```python
dbutils.widgets.text("catalog", "citibike_dev")
catalog = dbutils.widgets.get("catalog")
print(catalog)

dbutils.widgets.text("environment", "d")
environment = dbutils.widgets.get("environment")
print(environment)

if environment == "d":
    storage_credential = "skylink_dev_credential"
elif environment == "u":
    storage_credential = "skylink_uat_credential"
elif environment == "p":
    storage_credential = "skylink_prd_credential"
else:
    raise ValueError(
        f"Invalid environment '{environment}'. "
        "Expected one of: 'd', 'u', 'p'."
    )

print(storage_credential)
```

```python
# Create External Location using SQL
spark.sql(f"""
  CREATE EXTERNAL LOCATION IF NOT EXISTS skylink_dev_external
  URL 'abfss://skylink@skylink{environment}eastus2storage.dfs.core.windows.net/'
  WITH (STORAGE CREDENTIAL `skylink_dev_credential`)
  COMMENT 'External location for dev environment'
""")
```

```python
spark.sql(f"""
CREATE CATALOG IF NOT EXISTS {catalog} 
MANAGED LOCATION 'abfss://skylink@skylink{environment}eastus2storage.dfs.core.windows.net/{catalog}';
""")

spark.sql(f"""
CREATE SCHEMA IF NOT EXISTS {catalog}.00_landing;
""")

spark.sql(f"""
CREATE SCHEMA IF NOT EXISTS {catalog}.01_bronze;
""")

spark.sql(f"""
CREATE SCHEMA IF NOT EXISTS {catalog}.02_silver;
""")

spark.sql(f"""
CREATE SCHEMA IF NOT EXISTS {catalog}.03_gold
""")

spark.sql(f"""
CREATE VOLUME if NOT EXISTS {catalog}.00_landing.source_citibike_data          
          """)
```
